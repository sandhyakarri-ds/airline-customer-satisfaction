```python
#Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
#Loading the data
url=r"D:\archive\train.csv"
data=pd.read_csv(url)
print(data.head())
print('***************')
print(data.shape)
```

```python
data.info()
```

```python
data.describe()
```

```python
#checking missing values
data.isnull().sum()
```

```python
#Missing value imputation
data['Arrival Delay in Minutes']=data['Arrival Delay in Minutes'].fillna(data['Arrival Delay in Minutes'].median())
```

```python
#checking missing values again
data.isnull().sum()
```

```python
#Exploratory data analysis
#countplot of satisfaction
plt.figure(figsize=(7,5))
sb.countplot(x='satisfaction',palette='mako',data=data)
plt.xlabel('satisfaction level')
plt.ylabel('No.of passengers')
plt.title('Passengers vs satisfaction level')
plt.show()
```

```python
#Satisfcation by type of travel
plt.figure(figsize=(8,6))
sb.countplot(x='Type of Travel',hue='satisfaction',data=data,palette='crest')
plt.xlabel('Type of travel')
plt.ylabel('No.of passengers')
plt.title('Type of travel vs satisfaction')
plt.show()
```

```python
#Satisfaction by type of class
plt.figure(figsize=(7,5))
sb.countplot(x='Class',hue='satisfaction',data=data,palette='viridis')
plt.xlabel('class')
plt.ylabel('Satisfaction')
plt.title('Satisfaction by class')
plt.show()
```

```python
#Rating columns
ratings = data[['Inflight wifi service', 'Seat comfort', 'Food and drink', 'Online boarding', 'Inflight entertainment', 
                'On-board service', 'Leg room service', 'Baggage handling', 'Checkin service', 'Inflight service', 
                'Cleanliness']]

plt.figure(figsize=(10,8))
sb.heatmap(ratings.corr(), cmap='coolwarm',vmin=-1,vmax=+1)
plt.title('Correlation Between Service Ratings')
plt.show()
```

```python
#Cleanliness rating by class
plt.figure(figsize=(8,6))
sb.barplot(x='Class',y='Cleanliness',hue='satisfaction',data=data,palette='Reds')
plt.xlabel('Class')
plt.ylabel('Cleanliness')
plt.title('Cleanliness ratings by class')
plt.show()
```

```python
#Online boarding vs satisfaction
stack=pd.crosstab(data['Online boarding'],data['satisfaction'])
stack.plot(kind='bar',stacked='True',colormap='flare')
plt.title('Online boarding vs satisfaction')
plt.xlabel('Online boarding ratings')
plt.ylabel('Satisfaction')
plt.show()
```

```python
#Grouping
#Grouping by satisfaction rate
data['satisfaction'].value_counts(normalize=True)*100
```

```python
#Group by class satisfaction
data.groupby('Class')['satisfaction'].value_counts(normalize=True).unstack()*100
```

```python
#Average ratings
rating_cols = ['Inflight wifi service', 'Seat comfort', 'Food and drink',
               'Online boarding', 'Inflight entertainment', 'On-board service',
               'Leg room service', 'Baggage handling', 'Checkin service',
               'Inflight service', 'Cleanliness']

data.groupby('satisfaction')[rating_cols].mean().T.sort_values(by='satisfied', ascending=False)
```

```python
#satisfaction by arrival delay,departure delay
data.groupby('satisfaction')[['Departure Delay in Minutes','Arrival Delay in Minutes']].mean() 
```

```python
#Satisfaction by age group
bin=[0,20,35,50,65,80]
names=['Teen','Young adult','Adult','Middle age','Senior citizen']
data['Age']=pd.cut(data['Age'],bins=bin,labels=names)
data.groupby('Age')['satisfaction'].value_counts(normalize=True).unstack()*100
```

```python
data.to_csv(r'D:\archive\airline_cleaned.csv',index=False)
```
